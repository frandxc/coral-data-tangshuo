```python
import seaborn as sns
import matplotlib.pyplot as plt

# 使用 FacetGrid 绘制
g = sns.FacetGrid(data, col="species", col_wrap=4, sharex=False, sharey=False)

# 使用箱线图（Boxplot）代替散点图
g.map(sns.boxplot, "incu_temperature", "Total_photo", palette="Set2")

# 叠加数据点
g.map(sns.stripplot, "incu_temperature", "Total_photo", color="k", alpha=0.5, dodge=True)

# 设置标题
g.fig.suptitle("Boxplot by Species", y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from statannotations.Annotator import Annotator

# 确保温度组为字符串类型
data['incu_temperature'] = data['incu_temperature'].astype(str)

# 获取分类变量的顺序
order = sorted(data['incu_temperature'].unique())

# 使用 FacetGrid 绘制
g = sns.FacetGrid(data, col="species", col_wrap=4, sharex=False, sharey=False)

# 定义显著性对比组
pairs = [("27", "31"), ("31", "34"), ("27", "34")]

# 定义绘图函数，带显著性分析
def boxplot_with_significance(data, x, y, **kwargs):
    ax = plt.gca()  # 获取当前子图

    # 绘制箱线图，修复 `palette` 警告
    sns.boxplot(
        x=x,
        y=y,
        data=data,
        hue=x,  # 设置 `hue`，解决警告
        palette="Set2",
        order=order,
        ax=ax,
        dodge=False,
        legend=False,
    )

    # 绘制数据点
    sns.stripplot(
        x=x,
        y=y,
        data=data,
        hue=None,
        dodge=True,
        color="k",
        alpha=0.5,
        order=order,
        ax=ax,
    )

    # 添加显著性标注
    annotator = Annotator(ax, pairs, data=data, x=x, y=y, order=order)
    annotator.configure(test="t-test_ind", text_format="star", loc="outside", verbose=2)
    annotator.apply_and_annotate()

# 应用自定义函数绘制图像
g.map_dataframe(boxplot_with_significance, x="incu_temperature", y="Total_photo")

# 设置标题
g.fig.suptitle("Boxplot with Significance by Species", y=1.02)
plt.show()
```
